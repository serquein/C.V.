## Transformation of Categorical data to dummy variables

### Ordinal data - Label encoding

In this section I am encoding the ordinal data to dummy variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

#### ParentEduc

```python
df["ParentEduc"].unique()
```

```python
le.fit(["some high school","high school" ,"some college", "associate's degree", "bachelor's degree", "master's degree"])
df["ParentEduc"] = le.transform(df["ParentEduc"])
```

#### TestPrep

```python
df["TestPrep"].unique()
```

```python
le.fit(['none', 'completed'])
df["TestPrep"] = le.transform(df["TestPrep"])
```

#### PracticeSport

```python
df["PracticeSport"].unique()
```

```python
le.fit(["never", "sometimes", "regularly"])
df["PracticeSport"] = le.transform(df["PracticeSport"])
```

#### WklyStudyHours

```python
df["WklyStudyHours"].unique()
```

```python
le.fit(['< 5', '5 - 10', '> 10'])
df["WklyStudyHours"] = le.transform(df["WklyStudyHours"])
```

### Nominal - Getdummies

In this section I am encoding the nominal data to dummy variables

```python
df = pd.get_dummies(df, columns=["Gender", "EthnicGroup", "LunchType", "ParentMaritalStatus", "PracticeSport", "IsFirstChild", "TransportMeans"])
```

### Define base categories 
For each categorical variable I define a base category:</br>
**Gender:** male</br>
**EthnicGroup:** group A</br>
**LunchType:** standard</br>
**ParentMaritalStatus:** married</br>
**PracticeSport:** 0 (no sport)</br>
**IsFirstChild:** yes</br>
**TransportMeans:** school bus</br>

```python
df = df.drop(['Gender_male', 'EthnicGroup_group A', 'LunchType_standard', 'ParentMaritalStatus_married', 'PracticeSport_0', 'IsFirstChild_yes','TransportMeans_school_bus'] ,axis=1)
```

## Correlation map (Heatmap) 
I am checking the correlation between variables to select features for the analysis

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True 
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(16, 14))
    ax = sns.heatmap(corr, mask=mask,cmap='coolwarm', vmin=-1,vmax=1,annot=True, square=True)
```

### Delete features with high correlation
Reading and writing score have a correlation of 0.95 and therefore for my analysis I am keeping the reading score

```python
df = df.drop(['WritingScore'],axis=1)
```

# Linear Regression

```python
#Defining a random state to be able to replicate the same results
RAND_STATE = 34 

#Defining the test set size of 0.3
TT_RATIO = 0.3

#Splitting the Dataset in a train and test set
X=df.drop('MathScore',axis=1)
y=df['MathScore']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TT_RATIO, random_state=RAND_STATE)
```

## Model Development

```python
#Adding a constant
X_train_const_ct = sm.add_constant(X_train.to_numpy()) 

#Fit linear regression model
model = sm.OLS(y_train, X_train_const_ct).fit()
```

```python
#Print summary with feature names
features = list(['const']) + list(X.columns)
print_model = model.summary(xname=features)

print(print_model)
```

## Model Evaluation
Data scatter of predicted values of test dataset versus real values

```python
X_test_const_ct = sm.add_constant(X_test)

predictions_test = model.predict(X_test_const_ct)

plt.scatter(predictions_test, y_test, color = 'b')
plt.show()
```

# Decision Tree

## Model Development

### Optimal depth
I created a for loop to determine the optimal depth of the desicion tree base on the R squared

```python
col1 = []
col2 = []
for i in range (1,8):
    col1.append(i)

for i in col1:
    model = DecisionTreeRegressor(max_depth=i)
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test  = model.predict(X_test)
    r = model.score(X_test, y_test)
    col2.append(r)

DR = pd.DataFrame({'D':col1, 'R':col2})
DR
```

```python
max_index = DR['D'].idxmax()
key_max_number = DR.loc[max_index, 'D']
print('The optimal deepness is', key_max_number)

model = DecisionTreeRegressor(max_depth=7)
model.fit(X_train, y_train)
```

## Features importance
Here I am displaying the features ordered by their importance

```python
feature_names= list(X.columns)
importances = model.feature_importances_
features_importances = pd.Series(importances, index=feature_names)
features_importances

indices = np.argsort(importances)[::-1]
importances_sorted = importances[indices]
feature_names_sorted = np.array(feature_names)[indices]

plt.bar(feature_names_sorted, importances_sorted)
plt.xticks(rotation=90)
plt.title('Feature Importance')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.show()
```

## Model Evaluation
Calculating different evaluation metrics 

```python
#Comparing R squared of train and test set
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print(f'Training data accuracy: {train_accuracy}')
print(f'Testing data accuracy: {test_accuracy}')
```

```python
#Creating function to evaluate different metrics
y_pred_train = model.predict(X_train)
y_pred_test  = model.predict(X_test)

def performance_model(y_train, y_test, y_pred_train, y_pred_test):

    # Model validation y_train - y_pred_train
    ME_train = np.mean(y_train-y_pred_train)
    ME_test  = np.mean(y_test-y_pred_test)

    MAE_train = mean_absolute_error(y_train,y_pred_train)
    MAE_test  = mean_absolute_error(y_test,y_pred_test)

    MSE_train = mean_squared_error(y_train,y_pred_train)
    MSE_test  = mean_squared_error(y_test,y_pred_test)

    RMSE_train = np.sqrt(MSE_train)
    RMSE_test  = np.sqrt(MSE_test)

    MAPE_train = np.mean((np.abs(y_train-y_pred_train) / y_train)* 100.)
    MAPE_test  = np.mean((np.abs(y_test-y_pred_test) / y_test)* 100.)

    R2_train = r2_score(y_train,y_pred_train)
    R2_test  = r2_score(y_test,y_pred_test)

    performance = pd.DataFrame({'Error_metric': ['Mean error','Mean absolute error','Mean squared error',
                                             'Root mean squared error','Mean absolute percentual error',
                                             'R2'],
                            'Train': [ME_train, MAE_train, MSE_train, RMSE_train, MAPE_train, R2_train],
                            'Test' : [ME_test, MAE_test , MSE_test, RMSE_test, MAPE_test, R2_test]})

    pd.options.display.float_format = '{:.2f}'.format


    df_train = pd.DataFrame({'Real_value': y_train, 'Predicted_value': y_pred_train})
    df_test  = pd.DataFrame({'Real_value': y_test,  'Predicted_value': y_pred_test})

    return performance, df_train, df_test
```

```python
#Applying performance_model function 
performance, _ ,_ = performance_model(y_train, y_test, y_pred_train, y_pred_test)
performance
```
